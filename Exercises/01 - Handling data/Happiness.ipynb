## Data from World Happiness Report

The World Happiness Report is an annual publication of the United Nations Sustainable Development Solutions Network. It contains articles, and rankings of national happiness based on respondent ratings of their own lives, which the report also correlates with various life factors.

In this notebook we will explore the happiness of different countries and the features associated.
The datasets that we will use are available in *Data*: **happiness2020.pkl** and **countries_info.csv**.

Although the features are self-explanatory, here a summary: 

**happiness2020.pkl**
* country: *Name of the country*
* happiness_score: *Happiness score*
* social_support: *Social support (mitigation the effects of inequality)*
* healthy_life_expectancy: *Healthy Life Expectancy*
* freedom_of_choices: *Freedom to make life choices*
* generosity: *Generosity (charity, volunteers)*
* perception_of_corruption: *Corruption Perception*
* world_region: *Area of the world of the country*

**countries_info.csv**
* country_name: *Name of the country*
* area: *Area in sq mi*
* population: *Number of people*
* literacy: *Literacy percentage*

```python
import pandas as pd
%matplotlib inline

DATA_FOLDER = 'Data/'

HAPPINESS_DATASET = DATA_FOLDER+"happiness2020.csv"
COUNTRIES_DATASET = DATA_FOLDER+"countries_info.csv"
```

## Task 1: Load the data

Load the 2 datasets in Pandas dataframes (called *happiness* and *countries*), and show the first rows.


**Hint**: Use the correct reader and verify the data has the expected format.

```python
from IPython.display import display

# Load datasets
happiness = pd.read_csv(HAPPINESS_DATASET)
# Note: literacy uses comma as decimal separator
countries = pd.read_csv(COUNTRIES_DATASET, decimal=',')

# Show first rows
display(happiness.head())
display(countries.head())
```

## Task 2: Let's merge the data

Create a dataframe called *country_features* by merging *happiness* and *countries*. A row of this dataframe must describe all the features that we have about a country.

**Hint**: Verify that all the rows are in the final dataframe.

```python
#Make both country columns lowercase so they match
happiness['country'] = happiness['country'].str.lower()
countries['country_name'] = countries['country_name'].str.lower()

# Simple merge keeping all happiness rows
country_features = pd.merge(
    happiness,
    countries,
    left_on='country',
    right_on='country_name',
    how='left'
)

# Drop duplicate key from countries
country_features = country_features.drop(columns=['country_name'])

display(country_features.head())
```

## Task 3: Where are people happier?

Print the top 10 countries based on their happiness score (higher is better).

```python
from IPython.display import display

top10 = (
    happiness
    .sort_values('happiness_score', ascending=False)
    .loc[:, ['country', 'happiness_score']]
    .head(10)
)

display(top10)
```

We are interested to know in what world region people are happier. 

Create and print a dataframe with the (1) average happiness score and (2) the number of contries for each world region.
Sort the result to show the happiness ranking.

```python
from IPython.display import display

region_stats = (
    happiness
    .groupby('world_region', as_index=True)
    .agg(
        avg_happiness=('happiness_score', 'mean'),
        num_countries=('country', 'nunique')
    )
    .sort_values('avg_happiness', ascending=False)
    .reset_index()
)

# Ensure desired column order with region in the first column
region_stats = region_stats[['world_region', 'avg_happiness', 'num_countries']]

display(region_stats)
```

The first region has only a few countries! What are them and what is their score?

```python
# Countries and scores for the top region (region_stats is a DataFrame)
best_region = region_stats.iloc[0]['world_region']
subset = (
    happiness.loc[happiness['world_region'] == best_region, ['country', 'happiness_score']]
    .sort_values('happiness_score', ascending=False)
)
print(f"Top region: {best_region}")
subset
```

## Task 4: How literate is the world?

Print the names of the countries with a level of literacy of 100%. 

For each country, print the name and the world region in the format: *{region name} - {country name} ({happiness score})*

```python
# Countries with 100% literacy and their region with happiness score
fully_literate = (
    country_features.loc[country_features['literacy'] == 100, ['world_region', 'country', 'happiness_score']]
    .sort_values(['world_region', 'country'])
)

for _, row in fully_literate.iterrows():
    print(f"{row['world_region']} - {row['country']} ({row['happiness_score']:.3f})")
```

What is the global average?

```python
# Global average literacy (ignore missing values)
global_avg_literacy = country_features['literacy'].mean()
print(f"Global average literacy: {global_avg_literacy:.2f}%")
```

Calculate the proportion of countries with a literacy level below 50%. Print the value in percentage, formatted with 2 decimals.

```python
# Share of countries with literacy below 50%
num_countries_total = country_features['country'].nunique()
num_low_lit = country_features.loc[country_features['literacy'] < 50, 'country'].nunique()
share_low_lit = (num_low_lit / num_countries_total) * 100
print(f"Countries with literacy < 50%: {num_low_lit}/{num_countries_total} ({share_low_lit:.2f}%)")
```

Print the raw number and the percentage of world population that is illiterate.

```python
# Illiterate population total and percentage
# 'literacy' is a percentage; compute illiterate share per country
cf = country_features.dropna(subset=['population', 'literacy']).copy()
cf['illiterate_share'] = (100 - cf['literacy']) / 100
cf['illiterate_population'] = cf['population'] * cf['illiterate_share']

total_world_population = cf['population'].sum()
total_illiterate = cf['illiterate_population'].sum()
illiteracy_percent = (total_illiterate / total_world_population) * 100

print(f"Illiterate (count): {int(total_illiterate):,}")
print(f"Illiterate (percent of world population): {illiteracy_percent:.2f}%")
```
